```python
movie_matrix = df.pivot_table(index='user_id' , columns="title" , values = "rating")
```

```python
movie_matrix
```

```python
startwars = movie_matrix["Star Wars (1977)"]
```

```python
similar_to_starwars = movie_matrix.corrwith(startwars)
```

```python
corrstarwars = pd.DataFrame(similar_to_starwars,columns=["Correlation"])
```

```python
corrstarwars.dropna(inplace=True) #replace Nan value 
```

```python
corrstarwars.head()
```
